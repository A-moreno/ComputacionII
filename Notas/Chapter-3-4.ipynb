# Pilas (Stack)

Como explicamos en la primer clase las pilas son el tipo de dato más simple. Se trata de una lista donde los elementos más recientes serán los ultimos en ser procesados lo que se conoce como LIFO.
Las operaciones de un Stack se muestran a continuación.


- **Stack()** crea un nueva pila que esta vacia. No requiere argumentos.
- **push(item)** agrega un nuevo elemento al tope de la pila, y no regresa valor.
- **pop()** Elimina un elemento de la pila. La pila se modificará y se imprime el valor del elemento
- **isEmpty()** Regresa True si  la pila esta vacia, o False de lo contrario.
- **size()** Regresa el número de elementos en la Pila.
- **peek()** Regresa el elemento al tope de la pila pero no lo elimina.



size() returns the number of items on the stack. It needs no parameters and returns an integer.

````python
class Stack:
     def __init__(self):
         self.items = []

     def isEmpty(self):
         return self.items == []

     def push(self, item):
         self.items.append(item)

     def pop(self):
         return self.items.pop()

     def peek(self):
         return self.items[len(self.items)-1]

     def size(self):
         return len(self.items)
````

2. **Parentesis Balanceados**: Uno de los problemas de los lenguajes de programación es que antes de ejecutar un código, el compilador revisa si la expresión está bien escrita. Comenzando por las expresiones aritmeticas tales como

$$(5+6)*(7+8)/(4+3)$$

Donde los parentesis usados para las operaciones deben estar balanceados correctamente.

Utilizando un Stack podemos insertar los elementos y retirarlos uno a uno si al final nos queda algún elemento entonces no esta balanceada la expresión.


````python
def parChecker(symbolString):
    s = Stack()
    balanced = True
    index = 0
    while index < len(symbolString) and balanced:
        symbol = symbolString[index]
        if symbol == "(":
            s.push(symbol)
        else:
            if s.isEmpty():
                balanced = False
            else:
                s.pop()

        index = index + 1

    if balanced and s.isEmpty():
        return True
    else:
        return False

print(parChecker('((()))'))
print(parChecker('(()'))
````
